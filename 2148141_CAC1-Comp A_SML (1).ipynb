# ** Univariate Analysis of Continuous variables **

## `i) Histogram`


1. Histograms are used for **understanding the shape of the data distribution.**
2. It shows how frequently every value in a data set occurs in a relatively unbiased way or **how frequently a value falls into a particular bin.**

```python
data.hist(figsize=(10,10))
```

**For better understanding of each feature we'll look into each of the graphs seperately.**

### 1. Age of an individual

```python
data['age'].hist(bins=15,color='red')
plt.title("Histogram of age variable",fontsize=20)
plt.xlabel("Age of a person (in years)",fontsize=20)
plt.ylabel("Frequency of the number of people at a particular age.",fontsize=15)
```

### INTERPRETATION:

1. The number of modes in a data set can be estimated by the number of prominent peaks present in a histogram. 
    So, Age variable is **UNIMODAL**.
    
2. **It will be incorrect to say that the feature age is positively skewed. Instead we'll say that it is of uniform distribution with some unusual values.**

### 2. BMI of an individual

```python
data['bmi'].hist(bins=10)
plt.title("Histogram of BMI variable",fontsize=20)
plt.xlabel("BMI of a person",fontsize=20)
plt.ylabel("Frequency of the number of people at a particular BMI.",fontsize=15)
```

## INTERPRETATION:

**Feature BMI is UNIMODAL**

In terms of distribution there can be 2 cases:

1. Either positively skewed. If BMI is a weakly positively skewed then **Mode is slightly greater than 300**, median is slightly below 300 and mean is slightly below 200, because in positively skewed graph Mode > Median > Mode
2. Or normally distributed.

### So, for checking distribution we need to plot histogram along with frequency polygon using distplot.


### 3. Expenses of an individual

```python
data['expenses'].hist(color='black')
plt.title("Expenses Status")
```

### We're unable to get the exact count above each bin, so,
### Using Matplotlib library and adding frequency above each bin for Outlier detection.

```python
bins =15
arr=plt.hist(data['expenses'],bins,color='purple')
for i in range(bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]))
plt.title("Histogram for expenses",fontsize=15)
plt.xlabel("Expenses of a person (in $)",fontsize=15)
plt.ylabel("Frequency of the number of people at a some expense",fontsize=15)
plt.show()
```

### INTERPRETATION: Histograms are also used for outlier detection.

1. We are easily able to detect the outliers in the expenses column.
2. 4 individuals whose expenses are more than 50000 and are going upto 60000 are outliers in the expenses column.
3. This feature is **HIGHLY POSITIVELY SKEWED. And hence Mode > Median > Mean**

### 4. No. of children of an individual

```python
print(data['children'].unique())
```

**As there are 6 distinct values present in children column, so we are setting bins as 6.**

```python
bins = 6
arr = plt.hist(data['children'],bins,color = "red")
for i in range(bins):
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]),fontsize = 20)

plt.xlabel("No. of children an individual is having.",fontsize = 15)
plt.ylabel("No. of individuals.",fontsize = 15)
plt.title("Status of Number of children of an individual",fontsize = 25)
plt.show()
```

## Interpretation:

## `xi) Violin Plot`

```python
sns.set(style="whitegrid")
 
sns.violinplot(x="children", y="expenses", hue="smoker",
                   data=data, palette="Set2", dodge=True)
```

```python
sns.set(style="whitegrid")
 
sns.violinplot(x="children", y="bmi", hue="smoker",
                   data=data, palette="Set2", dodge=True)
```

### Interpretation:

It summarizes a data set using 6 measures: a boxplot is used to show the minimum, first quartile, median, third quartile, and maximum. There is also a probability density function (PDF), which is essentially a rotated, smoothed histogram. This is included to show the shape of the data set. Boxplots are similar to violin plots, but lack the probability density function. A wider PDF indicates that the value occurs more frequently, and a narrower density function indicates that the value occurs less frequently.

**Change in shapes of these violin plots are beacuse Expenses is skewed and bmi is normally distributed.**
